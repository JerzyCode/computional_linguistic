```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scratch_df = pd.read_csv("results/contrastive_model_training_metrics.csv")
scratch_df.head()
```

```python
train_losses = scratch_df["train_loss"]
# val_losses = scratch_df["val_loss"]
steps = [5 * i for i in range(len(train_losses))]

plt.figure(figsize=(6, 4))

plt.plot(steps, train_losses, label="Train Loss")
# plt.plot(steps, val_losses, label="Validation Loss")
plt.xlabel("Training Steps")
plt.ylabel("Loss")
# plt.title("Training and Validation Loss over Steps (Scratch Model)")
plt.title("Training Loss over Steps (Contrastive Model)")
plt.legend()
plt.grid()
plt.show()
```

```python
val_accuracy = scratch_df["val_accuracy"]
val_f1 = scratch_df["val_f1"]
steps = [10 * i for i in range(len(train_losses))]

plt.figure(figsize=(6, 4))

plt.plot(steps, val_accuracy, label="Validation Accuracy", c="blue")
plt.plot(steps, val_f1, label="Validation F1 Score", c="green")
plt.xlabel("Training Steps")
plt.ylabel("Score")
plt.title("Validation Accuracy and F1 Score over Steps (Scratch Model)")
plt.legend()
plt.grid()
plt.show()
```
