### Nothing to do with it, it is well prepared

```python
train_df, test_df = train_test_split(df, test_size=0.1, random_state=42, stratify=df["Class"])

train_df.to_csv("data/train.csv", index=False)
test_df.to_csv("data/test.csv", index=False)
```

```python
train_df.head(5)
```

```python
train_df["length"] = train_df["Text"].apply(lambda x: len(str(x).split()))

print("Min długość:", train_df["length"].min())
print("Max długość:", train_df["length"].max())
print("Średnia:", train_df["length"].mean())
print("Mediana:", train_df["length"].median())
```
